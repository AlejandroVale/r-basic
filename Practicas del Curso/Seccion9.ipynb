```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.0,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1,f(x1), 'ro',x2,f(x2),'k')
plt.subplot(2,1,2)
plt.plot(x2,f(x2),'g--')
plt.show
```

```python
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,1)
plt.plot(x1,f(x1), 'ro',x2,f(x2),'k')
plt.subplot(1,2,2)
plt.plot(x2,f(x2),'g--')
plt.show
```

```python
mu = 100
sigma = 20
x = mu + sigma * np.random.randn(10000)
```

```python
plt.figure(1,figsize = (20,6))
n, bins, patches = plt.hist(x, 50,facecolor='g', alpha=0.6)
plt.xlabel("Cociente Intelectual")
plt.ylabel("Probabilidad")
plt.title("Histogramas de CI")
plt.text(40,500,'$\\mu =100. \\ \\sigma=20$')
plt.axis([20,180,0,700])
plt.grid(True)
plt.annotate("Maximo", xy=( 100,650),xytext = (120,680), arrowprops = dict(facecolor = "black",shrink = 0.05))
```
