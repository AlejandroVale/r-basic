```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

plt.figure(1,figsize = (15,6))
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X) 

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X,1,Y+1,facecolor='mediumpurple')

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X,-1,Y-1, where= Y-1>-1, facecolor='mediumpurple')
plt.fill_between(X,-1,Y-1, where= Y-1<-1, facecolor='salmon')

plt.show()
```

```python
# Ejercicio 3
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

plt.figure(1,figsize = (15,6))

n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 

for x, y in zip(X, Y1):     

        plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom')

for x, y in zip(X, -Y2):     

        plt.text(x, y - 0.15, '%.2f' % y, ha='center', va='bottom')  

plt.ylim(-1.25, +1.25)
```
