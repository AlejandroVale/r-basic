```python
from selenium import webdriver
from bs4 import BeautifulSoup

from os import path
import time
from datetime import datetime

from pandas import DataFrame, read_excel, concat

import re
import sys, configparser
```

```python
destination_dir = './'
last_file = 'inec_ipc_last'
incremental_file = 'inec_ipc_incremental'

url = "https://inec.gob.pa"
chromeOptions = webdriver.ChromeOptions()
##chromeOptions.add_argument("--headless")
browser = webdriver.Chrome('./driver/chromedriver.exe', options=chromeOptions)
browser.implicitly_wait(30)
browser.get(url)
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')

plot = soup.find_all(id='graficas_graficaIndicadoresImageMap')[0]
points = plot.find_all('area', {'shape': 'circle'})
```

```python
def monthToNum(shortMonth):
    return str({
    'Ene' : 1,
    'Feb' : 2,
    'Mar' : 3,
    'Abr' : 4,
    'May' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Ago' : 8,
    'Sep' : 9, 
    'Oct' : 10,
    'Nov' : 11,
    'Dic' : 12
    }[shortMonth])


df = []
today = datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S")

for p in points:
    string = p.get('alt')

    date, value = re.findall(r'([a-zA-Z]{3}\s\d{4}).*Valor:\s(.*)', string)[0]
    mes = date[:3]
    date = date.replace(mes, monthToNum(mes))
    date = datetime.strptime(date, '%m %Y')

    vals = [date, value, today]
    df.append(vals)

#df = pd.DataFrame(df, columns=['date', 'value', 'last_update_utc'])
df = DataFrame(df, columns=['date', 'value', 'last_update_utc'])
df.to_excel(f"{destination_dir}{last_file}.xlsx", index=None)
```
