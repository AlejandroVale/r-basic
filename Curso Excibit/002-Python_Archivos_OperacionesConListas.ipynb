```python
def classifier(value):
    measure = value['Girth (in)']
    if measure >15:
        return 'grande'
    elif measure > 10:
        return 'medio'
    else:
        return 'chico'

df['category'] = df.apply(classifier, axis=1)

df
```

```python
df.groupby(['category']).mean()
```
