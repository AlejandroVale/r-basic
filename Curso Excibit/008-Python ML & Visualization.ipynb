```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
filename = './data/002/housepricing-train.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
sns.distplot(df['LotArea'])
```

```python
sns.distplot(df['SalePrice'])
```

```python
df.describe()
```

```python
data = pd.concat([df['SalePrice'], df['YearBuilt']], axis=1)
data.plot.scatter(x='YearBuilt', y='SalePrice')
```

```python
corr = df.corr()

sns.set(font_scale=0.8)
plt.subplots(figsize=(10,10))
sns.heatmap(corr)
```

```python
corr['SalePrice'].sort_values(ascending=False)
```

```python
k = 10

cols = corr.nlargest(k, 'SalePrice').index
print(cols)
```

```python
filtered_corr = df[cols].corr()
sns.heatmap(filtered_corr, xticklabels=cols.values, yticklabels=cols.values, annot=True)
```

```python
data = pd.concat([df['SalePrice'], df['GrLivArea']], axis=1)
data.plot.scatter(x='GrLivArea', y='SalePrice')
```

```python
data = pd.concat([df['SalePrice'], df['OverallQual']], axis=1)
data.plot.scatter(x='OverallQual', y='SalePrice')
```

```python

sns.scatterplot(x='OverallQual', y='SalePrice', data=data)
```

```python
sns.boxplot(x='OverallQual', y='SalePrice', data=data)
```
