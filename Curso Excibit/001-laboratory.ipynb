```python
query = "SELECT * FROM `excibitlab.covidmexico_100` LIMIT 1000"

options = PipelineOptions()
options = {'project': 'numeric-water-289923',
           'runner': 'DataflowRunner',
           'region': 'us-east1',
           'temp_location': 'gs://2021-01-14-dataflowtesting'
          }
pipeline_options = beam.pipeline.PipelineOptions(flags=[], **options)

p = beam.Pipeline(runner=InteractiveRunner(), options=pipeline_options)
    
```

```python
bq_data = (
  p
  | 'QueryTable' >> beam.io.ReadFromBigQuery(query = query, project='numeric-water-289923', use_standard_sql=True)
)
```

```python
class FilterByEdad(beam.DoFn):
    def __init__(self, edad):
        self._edad = edad

    def process(self, element):
        if element['EDAD'] >= self._edad:
            yield element

filtered_data = bq_data | 'Filter > 40 años' >> beam.ParDo(FilterByEdad(40))
```

```python
ib.show(filtered_data)
```

```python
class FilterByEdad(beam.DoFn):
    def __init__(self, edad):
        self._edad = edad

    def process(self, element, edad):
        if edad is None:
            edad = self._edad
        if element['EDAD'] >= edad:
            yield element
            
avg_age = (
    bq_data 
    | 'lista edad' >> beam.Map(lambda data: data['EDAD'])
    | 'calculo promedio' >> beam.combiners.Mean.Globally()
)

filtered_data = (
    bq_data 
    | 'Filter > 40 años' >> beam.ParDo(FilterByEdad(avg_age),
                                       edad=beam.pvalue.AsSingleton(avg_age))
)
#.combiners.Mean.Globally()

ib.show(filtered_data)
```

```python
chars = p | 'Create chars' >> beam.Create(['# \n t c a C'])

plants = (
  p
  | 'Gardening plants' >> beam.Create([
      '# 🍓Strawberry\n',
      '# 🥕Carrot\n',
      '# 🍆Eggplant\n',
      '# 🍅Tomato\n',
      '# 🥔Potato\n',
  ])
  | 'Strip header' >> beam.Map(
      lambda text,
      chars: text.strip(chars),
      chars=beam.pvalue.AsSingleton(chars),
    )
)

ib.show(plants)
```
